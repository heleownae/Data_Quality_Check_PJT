## 2. Ïù∏ÏΩîÎî© (ÏàòÏûëÏóÖ)

```python
sample_df = df.copy()
sample_df.head()
```

```python
# ÎìúÎûçÌï† Ïª¨Îüº Í∏∞Ï§Ä: Îã§Ï§ëÏÑ†ÌÉù ÏûàÎäî Ïª¨Îüº, ÌåüÏ∫êÏä§Ìä∏ Í¥ÄÎ†® Ïª¨Îüº, Íµ∞ÏßëÏù¥ Ïï†Îß§Ìïú? Ïª¨Îüº(Age)
col_for_drop = ['Age', 'spotify_listening_device', 'music_Influencial_mood', 'music_lis_frequency', 'preferred_listening_content', 'music_expl_method', 'pod_lis_frequency', 'fav_pod_genre', 'preffered_pod_format', 'pod_host_preference', 'preffered_pod_duration', 'pod_variety_satisfaction']

sample_df2 = sample_df.drop(columns=col_for_drop)
sample_df2.head()
```

```python
sample_df2.describe(include='all')
```

```python
for col in sample_df2.columns:
    print("Í≥†Ïú†Í∞í Í∞úÏàò: ", df[col].nunique())
    print("Í∞í Í∞úÏàò: ", df[col].count())
    print(df[col].value_counts())
    print()
```

```python
# Gender

def get_gender(gnd):
    if gnd == 'Others':
        return 0
    elif gnd == 'Female':
        return 2
    else:
        return 1

sample_df2['Gender'] = sample_df2['Gender'].apply(get_gender)
sample_df2[['Gender']]
```

```python
# spotify_usage_period

def get_useperiod(prd):
    if prd == 'Less than 6 months':
        return 0.5
    elif prd == '6 months to 1 year':
        return 0.5
    elif prd == '1 year to 2 years':
        return 1
    else:
        return 2
    
sample_df2['spotify_usage_period'] = sample_df2['spotify_usage_period'].apply(get_useperiod)
sample_df2[['spotify_usage_period']]
```

```python
# spotify_subscription_plan

def get_plan(pld):
    if pld == 'Free (ad-supported)':
        return 0
    else:
        return 1
    
sample_df2['spotify_subscription_plan'] = sample_df2['spotify_subscription_plan'].apply(get_plan)
sample_df2[['spotify_subscription_plan']]
```

```python
# premium_sub_willingness

def get_will(wil):
    if wil == 'No':
        return 0
    else:
        return 1
    
sample_df2['premium_sub_willingness'] = sample_df2['premium_sub_willingness'].apply(get_will)
sample_df2[['premium_sub_willingness']]
```

```python
# preffered_premium_plan

def get_prefplan(pref):
    if pref == 'Student Plan-Rs 59/month':
        return 0
    elif pref == 'Individual Plan- Rs 119/ month':
        return 1
    elif pref == 'Duo plan- Rs 149/month':
        return 2
    else:
        return 3

sample_df2['preffered_premium_plan'] = sample_df2['preffered_premium_plan'].apply(get_prefplan)
sample_df2[['preffered_premium_plan']]
```

```python
# fav_music_genre

def get_genre(gen):
    if gen == 'Melody':
        return 5
    elif gen == 'classical':
        return 4
    elif gen == 'Pop':
        return 3
    elif gen == 'Rap':
        return 2
    elif gen == 'Electronic/Dance':
        return 1
    else:
        return 0

sample_df2['fav_music_genre'] = sample_df2['fav_music_genre'].apply(get_genre)
sample_df2[['fav_music_genre']]
```

```python
# music_time_slot

def get_time(tim):
    if tim == 'Morning':
        return 1
    elif tim == 'Afternoon':
        return 2
    else:
        return 3

sample_df2['music_time_slot'] = sample_df2['music_time_slot'].apply(get_time)
sample_df2[['music_time_slot']]
```
